```python
import rouskinhf
import numpy as np
import pandas as pd

pri_miRNA= pd.DataFrame.from_dict(rouskinhf.get_dataset('pri_miRNA'), orient='index')
human_mRNA = pd.DataFrame.from_dict(rouskinhf.get_dataset('human_mRNA'), orient='index')
pri_miRNA['dataset'] = ['pri-miRNA (N={:,})'.format(len(pri_miRNA))]*pri_miRNA.shape[0]
human_mRNA['dataset'] = ['human mRNA (N={:,})'.format(len(human_mRNA))]*human_mRNA.shape[0]
df = pd.concat([pri_miRNA, human_mRNA])
df['length'] = df['sequence'].apply(len)
df
```

```python
# plot both length distributions

import plotly.graph_objs as go
import plotly.express as px

# make bar thinner
fig = px.histogram(df, x="length", color="dataset", barmode="relative", range_x=[0,100*np.ceil(max(df['length'])/100)], nbins=int(np.ceil(max(df['length'])/100)))
fig.update_layout(
    title="Length distribution of pri-miRNA and human mRNA",
    xaxis_title="Length",
    yaxis_title="Count",
    legend_title="Dataset",
    bargap=0.3,
    font=dict(
        size=18,
    ),
    xaxis=dict(
        tickmode='linear',
        tick0=0,
        dtick=100
    ),
    width=800,  
    height=600,
    legend=dict(
        yanchor="top",
        y=0.99,
        xanchor="right",
        x=0.99
    ),
    paper_bgcolor="white",
    # plot_bgcolor="white",
)

# to pdf
import plotly.io as pio
pio.write_image(fig, 'images/length_distribution.pdf')
fig.show()
```
